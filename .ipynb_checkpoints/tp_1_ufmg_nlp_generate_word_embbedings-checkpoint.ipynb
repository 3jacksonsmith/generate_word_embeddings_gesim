# Import

```python
df = pd.read_csv('questions-words.txt', sep=" ", header=true)
```

```python
df.columns = ["first","second","third","fourth"]
df.head()
```

```python
print(df.to_string())
```

# Transform Data to Array


```python
df['first']
```

```python
#Create an empty list 
row_list = [] 
  
# Iterate over each row 
for index, rows in df.iterrows(): 
    # Create list for the current row 
    my_list =[ rows['first'].upper(), rows['second'].upper(), rows['third'].upper(), rows['fourth'].upper() ]
      
    # append the list to the final list 
    row_list.append(my_list) 
```

```python
print(type(row_list))
print(row_list[0])

row_test_list = []
row_test_list.append(row_list[0])
row_test_list.append(row_list[1])
row_test_list.append(row_list[2])
row_test_list.append(row_list[3])
row_test_list.append(row_list[4])
```

```python
import gensim, logging


# train word2vec on the two sentences
model = gensim.models.Word2Vec(row_test_list, min_count=1)
```
